## It Pays to get Paid

    Next, I wanted to understand how true the idea is that consumers really don't like paying for apps and it is better to just make it free to get faster growth and better enagegment. While this may be true, the data indicates that consumers actually have much stronger ratings for Paid apps as opposed to Free apps! 

    In the dataset, apps were given the categorization "Free" or "Paid." First, I calculated the mean and median for the Free apps.

```python
fear = pd.read_csv('googleplaystore.csv', encoding='latin')
var2 = fear.loc[lambda df: fear['Type'] == 'Free', 'Rating']
free_mean = var2.mean()
var2.mean()
#Free Apps, Mean:
```

```python
fear = pd.read_csv('googleplaystore.csv', encoding='latin')
var2 = fear.loc[lambda df: fear['Type'] == 'Free', 'Rating']
free_median = var2.median()
var2.median()
#Free Apps, Median:
```

    Then, I calculated the mean and median for the Paid apps.

```python
fear = pd.read_csv('googleplaystore.csv', encoding='latin')
var1 = fear.loc[lambda df: fear['Type'] == 'Paid', 'Rating']
paid_mean = var1.mean()
var1.mean()
#Paid Apps, Mean:
```

```python
fear = pd.read_csv('googleplaystore.csv', encoding='latin')
var1 = fear.loc[lambda df: fear['Type'] == 'Paid', 'Rating']
paid_median = var1.median()
var1.median()
#Paid Apps, Median:
```

    Finally, I plotted a double bar graph to illustrate how the Paid apps overrall got higher ratings than the Free apps and how that differed by mean and median. I added a grid to make the differences more obvious. While the differences are small, it does demonstrate that people who take out their wallets are much more likely to leave a better rating. 

```python
paid = [paid_mean, paid_median]
free = [free_mean, free_median]
index = ['Mean', 'Median']
df = pd.DataFrame({'Paid': paid,'Free': free}, index=index) 
ax = df.plot.bar(figsize=(8,10),ylim=(0,5), alpha=1, width=.5, rot=0)

ax.set_title('Paid vs Free Apps Ratings')
ax.set_ylabel('Rating')
ax.set_xlabel('App Type')
#i'm setting axis from 0 - 5, and then spread every .1
ax.yaxis.set_ticks(np.arange(0, 5, 0.10))
plt.legend(['Free', 'Paid'], loc='lower center')
plt.grid()
plt.show()

```

## How Tough is ALL of your Competition?

    But why are the Paid apps more likely to get higher ratings? This small difference made me curious about how clustered or spread out the reviews were in the marketplace. Could it be that perhaps Paid apps were better-developed applications on average whereas the quality among Free apps could vary more widely?

    Using the graph below, I was able to see that the majority of app reviews are clustered between 4.0 and 4.5 within the blue box plot. What was interesting is that there actually are a significant portions of apps that are rated well below 3.5 stars. It seems that poorly performing apps aren't considered outliers but rather very common on the Play Store.
