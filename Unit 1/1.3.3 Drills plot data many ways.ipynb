```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
#apply in pandas df first. calling data in seaborn. 
#seaborn is working
```

Choose one variable and plot that variable four different ways.

```python
var1 = pd.read_csv('seaborn_heart.csv')
sns.countplot(x="chol", data=var1)
```

```python
var1 = pd.read_csv('seaborn_heart.csv')
sns.countplot(x="sex", data=var1)
```

Choose two continuous variables, and plot them three different ways.

```python
#simple scattterplot
var1 = pd.read_csv('seaborn_heart.csv')
sns.relplot(x="age", y="chol", data=var1)
```

```python
#Line graph with with linear regression
var1 = pd.read_csv('seaborn_heart.csv')
sns.lmplot(x="age", y="chol", data=var1)
```

```python
#Line graph without Confidence Interval - St. Deviation
var1 = pd.read_csv('seaborn_heart.csv')
sns.relplot(x="age", y="chol", data=var1, kind="line",ci="sd")
```

Choose one continuous variable and one categorical variable, and plot them six different ways.
