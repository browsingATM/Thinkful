```python
#Poisson
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

poisson = np.random.poisson([5], 100)
plt.hist(poisson)
# Add a vertical line at the mean.
plt.axvline(poisson.mean(), color='r', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.

plt.axvline(poisson.mean() + poisson.std(), color='r', linestyle='dashed', linewidth=1)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(poisson.mean() - poisson.std(), color='r', linestyle='dashed', linewidth=1) 

plt.show()
```

```python
#the mean and st dev are helpful indicatos to show where the data is skewed--non-normal distributions
```

```python
#additionally

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

var1 = np.random.normal(5, .5, 100)
var2 = np.random.normal(10, 1, 100)

plt.axvline(var1.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(var1.mean() + var1.std(), color='r', linestyle='dashed', linewidth=1)
plt.axvline(var1.mean() - var1.std(), color='r', linestyle='dashed', linewidth=1)

plt.axvline(var2.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(var2.mean() + var2.std(), color='r', linestyle='dashed', linewidth=1)
plt.axvline(var2.mean() - var2.std(), color='r', linestyle='dashed', linewidth=1)

plt.hist(var1)
plt.hist(var2)
plt.show()
```

```python
#Because it's a normal data set it doesn't reveal too much--it's a pretty clear story but helpful for accuracy
```
