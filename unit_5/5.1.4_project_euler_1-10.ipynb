Problem 1: If we list all the natural numbers below 10 that are multiples of 3 or 5, we get 3, 5, 6 and 9. The sum of these multiples is 23.

Find the sum of all the multiples of 3 or 5 below 1000.


```python
multiples= []

for number in range(1000):
    if number % 3 == 0:
        multiples.append(number)
    if number % 5 == 0:
        multiples.append(number)
        
unique_multiples = set(multiples)
print('sum of the multiples: {}'.format(sum(unique_multiples)))
print(unique_multiples)
```
